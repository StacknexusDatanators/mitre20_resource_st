```python
import requests
import streamlit as st

# Set up the OpenWeather API endpoint and API key
url = "https://api.openweathermap.org/data/2.5/weather"
api_key = "your_api_key"

# Get the latitude and longitude from the user
lat = st.number_input("Enter latitude:")
lon = st.number_input("Enter longitude:")

# Make the API request
params = {"lat": lat, "lon": lon, "appid": api_key}
response = requests.get(url, params=params)

# Check if the API request was successful
if response.status_code == 200:
    # Extract the weather data from the response
    weather_data = response.json()

    # Display the weather data in Streamlit
    st.write(f"Weather for {lat}, {lon}:")
    st.write(f"Temperature: {weather_data['main']['temp']} K")
    st.write(f"Feels like: {weather_data['main']['feels_like']} K")
    st.write(f"Humidity: {weather_data['main']['humidity']} %")
    st.write(f"Pressure: {weather_data['main']['pressure']} hPa")
    st.write(f"Wind speed: {weather_data['wind']['speed']} m/s")
    st.write(f"Wind direction: {weather_data['wind']['deg']} degrees")
else:
    # Display an error message if the API request was not successful
    st.write("Error: Unable to retrieve weather data.")
```

```python
from pyproj import Proj, transform

inProj = Proj(init='epsg:3857')
outProj = Proj(init='epsg:4326')

world_lon1, world_lat1 = transform(outProj,inProj,-180,-85)
world_lon2, world_lat2 = transform(outProj,inProj,180,85)
cartodb = get_provider(CARTODBPOSITRON)

fig = figure(plot_width=800, plot_height=700,
             x_range=(world_lon1, world_lon2),
             y_range=(world_lat1, world_lat2),
             x_axis_type="mercator", y_axis_type="mercator")

fig.add_tile(cartodb)

show(fig)
```
